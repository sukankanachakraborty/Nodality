```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 40})
```

```python
topic = 'Costofliving'
```

```python
df = pd.read_csv('Clusters_'+topic+'.csv')
```

```python
import matplotlib

cmap = matplotlib.cm.get_cmap('turbo')

clusters = [2,0,1]
clrs = ['forestgreen','goldenrod','darkorchid']
cluster_lbls = ['Leaders','Funnelers','Receivers']

fig, ax = plt.subplots(1,1, figsize=(15,15))
for i in range(3):
    x = df.loc[df['Cluster']==clusters[i],'Inherent']
    y = df.loc[df['Cluster']==clusters[i],'Active']
    ax.scatter(x, y, c=clrs[i], label=cluster_lbls[i],
               alpha=0.7, edgecolors='none', s=500)
    
ax.legend(loc="lower left")
ax.grid(True)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')

plt.xlim([-10, 10])
plt.ylim([-4, 4])

plt.title('Cost of living')
plt.show()
plt.tight_layout()
```
